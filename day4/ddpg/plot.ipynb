```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = './eval_log/Pendulum-v0/DDPG_0831-210609.csv'
eval_history = np.genfromtxt(path, delimiter=',')
print(eval_history.shape)
```

```python
steps, scores = (eval_history[:, 0] / 1000.), eval_history[:, 1]
window_width = 5
smoothed_scores = np.zeros(len(scores))
for i in range(len(scores)):
    start = max(0, i - window_width)
    smoothed_scores[i] = np.mean(scores[start : i + 1])
```

```python
plt.plot(steps, smoothed_scores, color='tab:red')

plt.xlabel(r'step ($\times 10^3$)')
plt.ylabel('score')
plt.grid()
plt.xlim(steps[0], steps[-1])
plt.title('Pendulum-v0 (DDPG)')

plt.savefig('plot.png')
```
